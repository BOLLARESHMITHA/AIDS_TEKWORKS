```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
months=['jan','feb','mar','apr','may','jun','jul']
sales=[300,400,150,400,600,200,120]
plt.plot(months,sales,color='blue',marker='o',linestyle='-.')
plt.title("year sales")
plt.xlabel("months")
plt.ylabel("sales in dollor ")
plt.grid(True)
plt.show()
```

```python
prod=['TV','LAPTOP','MOBILE','TABLET']
revenue=[40000,60000,20000,30000]
plt.bar(prod,revenue,color='blue')
plt.title("revenue by products ")
plt.xlabel("products")
plt.ylabel("revenue ")
plt.grid(True)
plt.show()
```

```python
age=[22,25,30,35,40,45,50]
sal=[25000,28000,35000,40000,48000,50000,52000]
plt.scatter(age,sal,color='blue')
plt.title("age vs salary relationship ")
plt.xlabel("age")
plt.ylabel("salary in dollar")
plt.grid(True)
plt.show()
```

```python
brands=['Apple','Samsung','OnePlus','Other']
market_share=[30,40,10,20]
colors=['orange','grey','green','pink']
plt.pie(market_share,labels=brands,colors=colors,autopct='%1.1f%%',startangle=90)
plt.title("moblie market share")
```

```python
ages=np.random.randint(18,60,100)
plt.hist(ages,bins=8,color='lightblue',edgecolor='black')
plt.title("age distribution of emp")
plt.xlabel("age grp")
plt.ylabel("count of emps")
plt.show()
```

```python
x=[1,2,3,4,5]
y1=[10,20,25,30,35]
y2=[5,15,20,25,30]
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.plot(x,y1,'r--')
plt.title("line plot")
plt.subplot(1,2,2)
plt.bar(x,y2,color='lightblue')
```

```python
x=[1,2,3,4,5]
sales=[10,20,30,25,35]
profit=[2,5,8,6,10]
plt.figure(figsize=(10,3))
plt.plot(x,sales,label="sales",marker='o')
plt.plot(x,profit,label='profit',marker='s')
plt.title("sales vs profit")
plt.xlabel("quarter")
plt.ylabel("amt in dollor ")
plt.legend()
plt.savefig('sales_profit_chart.png')
```
