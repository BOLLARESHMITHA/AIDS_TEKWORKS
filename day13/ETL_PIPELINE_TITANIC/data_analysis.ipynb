✔ Cleaning Tasks
Convert "TotalCharges" to numeric (dataset has spaces → become NaN).
Fill missing numeric values using:
Median for tenure, MonthlyCharges, TotalCharges.
Replace missing categorical values with "Unknown".
✔ Feature Engineering
Create the following new columns:
1. tenure_group
Based on tenure months:
0–12   → "New"
13–36  → "Regular"
37–60  → "Loyal"
60+    → "Champion"
2. monthly_charge_segment
MonthlyCharges < 30  → "Low"
30–70              → "Medium"
> 70                 → "High"
3. has_internet_service
Convert InternetService column:
"DSL" / "Fiber optic" → 1
"No" → 0
4. is_multi_line_user
1 if MultipleLines == "Yes"
0 otherwise
5. contract_type_code
Map:
Month-to-month → 0
One year      → 1
Two year      → 2
✔ Drop unnecessary fields
Remove:
customerID, gender

```python
df['TotalCharges']=df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
def tenure_category(x):
    if x <= 12:
        return 'New'
    elif x <= 36:
        return 'Regular'
    elif x <= 60:
        return 'Loyal'
    else:
        return 'Champion'

df['tenure_group'] = df['tenure'].apply(tenure_category)
```

```python
def mon_charge(t):
    if(t<30):
        return 'Low'
    elif(t>=30 and t<=70):
        return 'Medium'
    else:
        return "High"
df['monthly_charge_segment']=df['MonthlyCharges'].apply(mon_charge)
```

```python
def has_net(t):
    if(t=='DSL' or t=='Fiber optic'):
        return 1
    else:
        return 0
df['has_internet_service']=df['InternetService'].apply(has_net)
```

```python
def multi_line(t):
    if(t=='Yes'):
        return 1
    else:
        return 0
df['is_multi_line']=df['MultipleLines'].apply(multi_line)
```

```python
def contract_type(x):
    if x=='Month-to-month':
        return 0
    elif x=='One year':
        return 1
    elif x=='Two year':
        return 2
df['contract_type_code']=df['Contract'].apply(contract_type)
```
